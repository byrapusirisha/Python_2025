**Global variable vs local variable**

- Varaibles outside the function is called as **global variable**

   - Global variables can be used anywhere
   - Global variabels can be used in inside or outside of the function
  
- Varaibles inside the function is called as **local varaible**

   - local variables can be used only inside function
   - local varaibled can not be used outside the function
